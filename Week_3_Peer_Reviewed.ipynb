### 3. In the next section we'll read geospatial data to help build latitude and longitude into our dataframe. We'll finish by mapping this new data frame onto the city of toronto by using the folium data.

```python
geodata = pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
geodata
```

```python
WrangledWithGeodata = pd.merge(wrangled, geodata, on="Postal Code")
WrangledWithGeodata['Borough'].nunique()
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(WrangledWithGeodata['Latitude'], WrangledWithGeodata['Longitude'], WrangledWithGeodata['Borough'], WrangledWithGeodata['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

### 4. Finally, we'll deploy the KMeans clustering method to understand what a euclidean based density approach to alternative postal codes might look like, if we believe partitioning by physical density to be the most equitable method of choice.

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# set number of clusters
kclusters = 10

torontoloc = WrangledWithGeodata.drop(['Neighborhood','Borough','Postal Code'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=1).fit(torontoloc)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
# add clustering labels
WrangledWithGeodata.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(WrangledWithGeodata['Latitude'], WrangledWithGeodata['Longitude'], WrangledWithGeodata['Neighborhood'], WrangledWithGeodata['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
